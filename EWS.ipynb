```python
from telegram import Update
from telegram.ext import Application, CommandHandler, CallbackContext
import nest_asyncio
import asyncio

# Terapkan nest_asyncio agar bisa menjalankan event loop dalam lingkungan yang sudah ada
nest_asyncio.apply()

# Konfigurasi
BOT_TOKEN = "7663919768:AAF3vGrRHwqA6oBKC5UENbcF-TnVefiyyPo"
GROUP_ID = "-2319713006"

# Fungsi untuk merespons perintah /start
async def start(update: Update, context: CallbackContext):
    user = update.effective_user
    await update.message.reply_text(f"Hello, {user.first_name}! I'm your Telegram bot. How can I assist you?")

    # Kirim pesan ke grup
    await context.bot.send_message(chat_id=GROUP_ID, text=f"User {user.first_name} started the bot!")

# Fungsi untuk merespons perintah /stop
async def stop(update: Update, context: CallbackContext):
    await update.message.reply_text("Bot stopped. See you next time!")
    # Menambahkan log ke grup jika bot dihentikan
    await context.bot.send_message(chat_id=GROUP_ID, text="The bot was stopped.")

# Fungsi utama untuk menjalankan bot
async def main():
    # Membuat instance dari aplikasi
    application = Application.builder().token(BOT_TOKEN).build()

    # Menambahkan handler untuk perintah /start dan /stop
    application.add_handler(CommandHandler("start", start))
    application.add_handler(CommandHandler("stop", stop))

    # Mulai bot
    await application.run_polling()

if __name__ == '__main__':
    # Menjalankan bot dengan asyncio
    asyncio.run(main())
```
